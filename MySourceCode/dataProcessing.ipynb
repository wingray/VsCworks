```python
def takeSampleName (df_input):
    X_tp=[]
    for index,str in enumerate(df_input):
        str=re.search(r'(?<=\\)[^\\]+(?=-)',str).group()
        X_tp=X_tp+[str]
    df_input=pd.DataFrame(X_tp)
    return df_input
```
